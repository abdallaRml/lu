```
import pandas as pd
import requests
from google.colab import files
from bs4 import BeautifulSoup
import time

# スクレイプするサイトの指定と「次へ」の最終ページ数・データフレーム指定！
base_url = 'https://design.style4.info/'
max_page = 128
cols = ['item', 'url']
df = pd.DataFrame(columns = cols)

# 最終ページまで繰り返し処理
for i in range(1, max_page - 124):
    print('-' * 70)
    
    # 文字化け防止とデータ解析
    response = requests.get(base_url + '/page/' + str(i))
    response.encoding = response.apparent_encoding
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 記事の１つのカタマリをリスト化でもってくる
    item_list = soup.find_all('div', {'class': 'item-main'})
    # 記事のカタマリを１つずつ持ってくるんだよォォーーーーーーー！
    for item in item_list:
        
        # GoogleAdsの回避
        if item.a is None:
            pass
        else:
            #print(item)
            url = item.h2.a.get('href')
            name = item.h2.a.string
            print(name)
            print(url)
            se = pd.Series([name, url], cols)
            df = df.append(se, ignore_index=True)
    time.sleep(1)

df.info()
filename = 'Fun_article.csv'
df.to_csv(filename, encoding = 'utf-8-sig')
files.download(filename)
```
